# PBMC3K labels
The data consist of 3k PBMCs from a Healthy Donor and are freely available from 10x Genomics. The following steps grabs the cell type clustering labels as assigned in the Seurat guided clustering tutorial.

```r
pbmc <- readRDS(file = "data/pbmc3k_final.rds")
```

```r
DimPlot(pbmc, reduction = "umap")
```

```r
pbmc$cell_type <- Idents(pbmc)
```

```r
pbmc[[]]
```

```r
write.table(x=pbmc[[]], file = "data/pbmc3k_SeuratMetadata.tsv", sep = "\t", quote = FALSE)
```
